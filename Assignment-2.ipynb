```python
data={'English':{'Sam':60,'Jackson':74,'Ahree':85}, 'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}}
series_data = {}
for subject, scores in data.items():
    for student, score in scores.items():
        series_data[(subject, student)] = score
```

```python
series = pd.Series(series_data)
print(series)
```

```python
data_df = pd.DataFrame(series,columns=["Final Scores"]).unstack(fill_value=0)


print(data_df)
```
