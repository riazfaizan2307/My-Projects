```python
import math
import pandas as pd
import numpy as np
import sklearn
import scipy.stats as stats
```

## Problem-1

```python
def calculate_confidence_interval(x):
    # Sample size
    n = 200

    p_hat = x / n

    confidence_level = 0.90

    # Calculate the standard error
    standard_error = math.sqrt((p_hat * (1 - p_hat)) / n)

    # Find the critical value for the confidence interval
    z_critical = stats.norm.ppf((1 + confidence_level) / 2)

    # Calculate the margin of error
    margin_of_error = z_critical * standard_error

    # Calculate the confidence interval
    lower_bound = p_hat - margin_of_error
    upper_bound = p_hat + margin_of_error

    return round(lower_bound, 5), round(upper_bound, 5), round(margin_of_error, 5)

x = int(input("Enter the number of highly satisfied customers: "))

lower_bound, upper_bound, margin_of_error = calculate_confidence_interval(x)

print(f"Confidence Interval: [{lower_bound}, {upper_bound}]")
print(f"Margin of Error: {margin_of_error}")
```

## Problem-2

```python
def probability_more_than_x(speed_x, mean, std_dev):
    z_score = (speed_x - mean) / std_dev

    probability = 1 - stats.norm.cdf(z_score)

    return round(probability, 4)

mean_speed = 75  # Mean speed in km/hr
std_dev_speed = 15  # Standard deviation in km/hr

speed_x = float(input("Enter the speed (X) in km/hr: "))

result_probability = probability_more_than_x(speed_x, mean_speed, std_dev_speed)

print(f"The probability that a car is traveling at more than {speed_x} km/hr is: {result_probability}")
```

```python
from sklearn.datasets import load_wine

data = load_wine()

wine_df = pd.DataFrame(data.data, columns=data.feature_names)
wine_df['target'] = data.target 
```

```python
wine_df
```

```python
from sklearn.model_selection import train_test_split
random_state = 100
sample_size = 50

_, sample_df, _, _ = train_test_split(wine_df, wine_df['target'], test_size=sample_size / len(wine_df), random_state=random_state)
```

```python
sample_mean = sample_df['alcohol'].mean()
sample_std = sample_df['alcohol'].std()

confidence_level = 0.95

z_critical = stats.norm.ppf(confidence_level + (1 - confidence_level) / 2)

margin_of_error = z_critical * (sample_std / (sample_size ** 0.5))

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Z-critical value: {z_critical:.3f}")
print(f"Margin of Error: {margin_of_error:.3f}")
print(f"Confidence Interval: {confidence_interval}")
```
