```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
df1 = pd.read_csv(r"C:\Users\HP\Documents\ALL EDUREKA PRACTICE CODES\All Data Files\Consumo_cerveja.csv")
df1.head(50)
```

```python
df1.describe()
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
df1["Temperatura Media (C)"] = df1["Temperatura Media (C)"].str.replace(",", ".").astype(float)
df1["Temperatura Minima (C)"] = df1["Temperatura Minima (C)"].str.replace(",", ".").astype(float)
df1["Temperatura Maxima (C)"] = df1["Temperatura Maxima (C)"].str.replace(",", ".").astype(float)
df1["Precipitacao (mm)"] = df1["Precipitacao (mm)"].str.replace(",", ".").astype(float)
```

```python
df1.info()
```

```python
df1["Month"] = df1["Data"].str.split("-" , expand=True)[1]
df1["Date"] = df1["Data"].str.split("-" , expand=True)[2]
df1 = df1.astype({"Month": float, "Date": float})
```

```python
df1.dropna(inplace=True)
df1.drop_duplicates(inplace=True)
df1.drop(columns = ["Data"], inplace=True)
```

```python
df1
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size
sns.boxplot(data=df1[['Temperatura Media (C)', 'Temperatura Minima (C)', 'Temperatura Maxima (C)', 'Precipitacao (mm)']])
plt.title('Outlier Analysis', fontsize=16)  # Increase the title font size
plt.xlabel('Features', fontsize=14)  # Increase the x-axis label font size
plt.ylabel('Values', fontsize=14)  # Increase the y-axis label font size
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
df1.info()
```

```python
df1.describe()
```

```python
Q1 = df1['Precipitacao (mm)'].quantile(0.25)
Q3 = df1['Precipitacao (mm)'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds to detect outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
filtered_df1 = df1[(df1['Precipitacao (mm)'] >= lower_bound) & (df1['Precipitacao (mm)'] <= upper_bound)]
```

```python
corr = df1.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
X = df1.drop('Consumo de cerveja (litros)', axis=1)
y = df1['Consumo de cerveja (litros)']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)
```

```python
print("Shape of the train dataset features")
print(X_train.shape)
print("Shape of the test dataset features")
print(X_test.shape)
```

```python
print("Shape of the train dataset target")
print(y_train.shape)
print("Shape of the test dataset target")
print(y_test.shape)
```

```python
model_linreg = LinearRegression()
model_linreg
```

```python
model_linreg.fit(X_train,y_train) 
```

```python
print("The Intercept for this Multiple Linear Regression model on the Consumo_cerveja dataset is ",
    model_linreg.intercept_)
```

```python
print("The Regression Co-efficient for this Multiple Linear Regression model on the Consumo_cerveja dataset are ")
print("====================================================================")
print(model_linreg.coef_)
```

```python
y_train_pred = model_linreg.predict(X_train) #training dataset prediction
```

```python
y_train_pred
```

```python
print("The RMSE value of the MLR model on the train dataset is: ")
print(np.sqrt(mean_squared_error(y_train, y_train_pred))) #actual values,predicted values
```

```python
print("The R2 score of the MLR model on the train dataset is:")
print(model_linreg.score(X_train,y_train))
```

```python
print("The mean absolute error of the MLR model on the train dataset is:")
print(mean_absolute_error(y_train, y_train_pred))
```

```python
y_test_pred = model_linreg.predict(X_test)
```

```python
print(y_test_pred)
```

```python
actual_vs_pred_df = pd.DataFrame({'Actual': y_test, 'Predicted' : y_test_pred})
```

```python
actual_vs_pred_df
```

```python
print("The RMSE value of the MLR model on the test dataset is: ")
print(np.sqrt(mean_squared_error(y_test, y_test_pred)))
```

```python
print("The R2 score of the MLR model on the test dataset is:")
print(model_linreg.score(X_test,y_test))
```

```python
print("The mean absolute error of the MLR model on the test dataset is:")
print(mean_absolute_error(y_test, y_test_pred))
```
